### Why is the error $\Delta(S^*) = 0.125$ ?

The optimal approximation error for a convex univariate function is calculated using the formula:

$$
\Delta(S^*) = \frac{c-a}{2} \left[f'(c) - f'(d)\right]
$$

#### Given
1. The function is $f(x) = x^2$, defined over the interval $[0, 1]$.
2. The derivative of the function is $f'(x) = 2x$.

We aim to calculate the **optimal approximation error** $\Delta(S^*)$ for the function $f(x) = x^2$ over the interval $[0, 1]$.

---

#### Step 1: Compute \( c \)

From the formula for \( c \):

$$
f'(c) = \frac{f(b) - f(a)}{b-a}
$$

Substitute the values:
- $f(b) = 1^2 = 1$
- $f(a) = 0$
- $b - a = 1 - 0 = 1$

Thus:

$$
f'(c) = \frac{1 - 0}{1} = 1
$$

Since $f'(x) = 2x$, we have:

$$
2c = 1 \quad \Rightarrow \quad c = 0.5
$$

---

#### Step 2: Compute $d$

From the formula for $d$:

$$
f'(d) = \frac{f(c) - f(a)}{c-a}
$$

Substitute the values:
- $f(c) = (0.5)^2 = 0.25$
- $f(a) = 0$
- $c - a = 0.5 - 0 = 0.5$

Thus:

$$
f'(d) = \frac{0.25 - 0}{0.5} = 0.5
$$

---

#### Step 3: Compute \( \Delta(S^*) \)

Using the formula for \( \Delta(S^*) \):

$$
\Delta(S^*) = \frac{c-a}{2} \left[f'(c) - f'(d)\right]
$$

Substitute the values:
- $c - a = 0.5 - 0 = 0.5$
- $f'(c) = 1$
- $f'(d) = 0.5$

Thus:

$$
\Delta(S^*) = \frac{0.5}{2} \left[1 - 0.5\right]
$$

Simplify:

$$
\Delta(S^*) = \frac{0.5}{2} \cdot 0.5 = 0.125
$$

---

### Conclusion

The optimal approximation error is:

$$
\Delta(S^*) = 0.125
$$

This result is derived from the specific properties of the function $f(x) = x^2$ and the optimal conditions for piecewise linear approximation over the interval $[0, 1]$.


```python
n_segments = 2
segment_intervals = np.linspace(a, b, n_segments + 1)

# Compute (S*) for each segment
errors = []
for i in range(n_segments):
    interval = (segment_intervals[i], segment_intervals[i+1])
    error = compute_approximation_error(convex_function, convex_derivative, interval)
    errors.append(error)

# Print the computed errors for all segments
for i, error in enumerate(errors, start=1):
    print(f"(S*_{i}) = {error:.6f}")

# Check if all errors are equal
if all(np.isclose(errors[i], errors[0]) for i in range(1, len(errors))):
    print("\nCondition verified: (S*_1) = (S*_2) = ... = (f*_n)")
else:
    print("\nCondition NOT satisfied: Errors are not equal.")
```

Lemma 2: For convex functions $f(x)$, if one enlarges $[a, b]$ by moving the endpoint $b$ to the right, or $a$ to the left, the optimal approximation error $\Delta(S^*)$ of the function $f(x)$ by a line segment S(x) will increase. On the contrary, if one reduces the interval $[a, b]$ by moving the endpoint $b$ to the left, or $a$ to the right, the optimal approximation error $\Delta(S^*)$ will decrease.

```python
a, b = -1, 2 # move b to the right, and a to the left
error = compute_approximation_error(convex_function, convex_derivative, interval=(a, b))
print(f"\nTheoretical Approximation Error: {error}, larger than before")
```
